```python
from ultralytics import YOLO
# Load a pretrained YOLOv8 model
model = YOLO('yolov8n.pt')
```

```python
results = model.predict("/content/who.jpg", conf=0.5)  # Set confidence threshold to 0.5
```

```python
model = YOLO('yolov8l.pt')
```

```python
results = model.predict(source="/content/peep.jpeg", conf=0.3, save=True)
```

```python
import cv2
from matplotlib import pyplot as plt
output_img = cv2.imread("runs/detect/predict/peep.jpeg")
plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
